```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import pycountry
import numpy as np
```

```python
df = pd.read_excel('ProductionCapa-Company.xlsx', 'years')
df
```

```python
df = df.drop(columns=
             ['Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17'],
             axis=1)
df
```

```python
df = df.drop('2023 (market share)', axis=1)
df
```

```python
df.info()
```

```python
print(df.isna().sum())
```

```python
def convert_alpha3(df=pd.DataFrame):
    alpha3 = []
    others = []

    for land in df["Country"]:
        try:
            alpha3.append(pycountry.countries.get(name=land).alpha_3)
        except:
            others.append(land)

    # print(set(alpha3))
    # print(set(others))
    
    return alpha3, others
```

```python
alpha3, others = convert_alpha3(df)

print(set(alpha3))
print(set(others))
```

```python
clean_df = df.copy()
others = []

for country in clean_df["Country"]:
    if country == "North America, Mexico":
        clean_df.replace(to_replace=country, value="United States of America", inplace=True)
    elif country == "South Corea":
        clean_df.replace(to_replace=country, value='Republic of Korea', inplace=True)
    else:
        others.append('land')
```

```python
alpha3_clean, others_clean = convert_alpha3(clean_df)

print(set(alpha3_clean))
print(set(others_clean))
```

```python
i = 0
for country in clean_df['Country']:
    if country == 'United States of America':
        clean_df.at[i, 'alpha3_code']='USA'
        i += 1
    elif country == 'Republic of Korea':
        clean_df.at[i, 'alpha3_code']='ROK'
        i += 1
    else:
        clean_df.at[i, 'alpha3_code']= str((pycountry.countries.get(name=country).alpha_3))
        i += 1
```
